```python
#Decorators

def main_welcome(func):
    
    def sub_welcome():
        print("Hello")
        func()
        print("How are you")
    return sub_welcome()

def intro_again():
    print("Hi Dude")
```

```python
intro_again()
```

```python
main_welcome(intro_again)
```

```python
@main_welcome
def intro_again():
    print("Hi Dude")
```

```python
def my_decorator(func):
    def wrapper():
        print("Hello")
        func()
        print("How are you")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hi Dude")
```

```python
say_hello()
```

```python
#Decorators with arguments

def repeat(n):
    def decorator(func):
        def wrapper(*args,**kwargs):
            for _ in range(n):
                func(*args,**kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("Hello")
```

```python
say_hello()
```

```python
class Employee:
    company="HP"

    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    
    #Instance Method (default)
    def print_info(self):
        info = f"Name : {self.name} , Salary : {self.salary}"
        print(info)

    #static method
    @staticmethod
    def sum(a,b):
        return a+b
    
    #Class method
    @classmethod
    def print_company(cls):
        print(cls.company)

    @classmethod
    def change_company(cls, new_company):
        cls.company=new_company


e1=Employee("Shubham",53000)
e2=Employee("Raunak",35000)

print(Employee.company)
e1.print_info()
e2.print_info()
```

```python
print(e1.sum(2,3))
e1.print_company()
e1.change_company("Acer")
e1.print_company()
```
