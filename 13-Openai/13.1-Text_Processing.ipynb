```python
from openai import OpenAI
client= OpenAI()

instruction = """Explain what this Python code does in one sentence:
number = int(input("Enter a number:"))
primeFlag = True
for i in range(2, number):
    if number % i == 0:
        primeFlag = False
        break
if primeFlag == True:
    print(number, " is prime.")
else:
    print(number, " is not prime.")
"""

response= client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {
            "role":"system",
            "content":[
                {
                    "type":"text",
                    "text":"You are a python tutor"
                }
            ]
        },
        {
            "role":"user",
            "content":[
                {
                    "type":"text",
                    "text":instruction
                }
            ]
        }
    ],
    response_format={
        "type": "text"
    },
    max_tokens=300,
)

print(response.choices[0].message.content)
```

```python
from openai import OpenAI
client= OpenAI()

response= client.responses.create(
    model="gpt-5.4-mini",
    input=[
        {
            "role":"developer",
            "content":"You are a python tutor"
        },
        {
            "role":"user",
            "content":"What does the sum() function does ?"
        },
        {
            "role":"assistant",
            "content":"The sum() function returns the sum of all the numbers in the list"
        },
        {
            "role":"user",
            "content":"What does the len() function does ?"
        }
    ]
)

print(response.output_text)
```

```python
from openai import OpenAI
client= OpenAI()

message= [{"role":"developer", "content":"You are a math tutor"}]
user_input=["Explain what is pi","Summarize your previous explantion in two bullet points","Translate those bullet points to hindi"]

for each_input in user_input:
    user_dict={'role':'user','content':each_input}
    message.append(user_dict)
    response= client.responses.create(
        model="gpt-5.4-mini",
        input=message)
    assistant_dict={'role':'assistant','content': response.output_text}
    message.append(assistant_dict)
    print("\n" + "="*50)
    print(response.output_text)
```
