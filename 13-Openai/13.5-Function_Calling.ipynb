```python
from openai import OpenAI
import json
client= OpenAI()


def get_stock_price(symbol):
    """Get the current stock price for given symbol"""
    stock_info = {
        "symbol": symbol,
        "price": "1000"
    }
    return json.dumps(stock_info)


tools = [
    {
        "type":"function",
        "name": "get_stock_price",
        "description": "Get the current stock price for the given symbol",
        "parameters": {
            "type": "object",
            "properties": {  
                "symbol": {
                    "type": "string",
                    "description": "The symbol, e.g. AAPL",
                }
            },
            "required": ["symbol"],
        },
    }
]


response= client.responses.create(
    model="gpt-5.4-mini",
    input="Get stock symbol and price for Apple Inc",
    tools=tools
)

print(response)
print(response.output)

```

```python
from openai import OpenAI
import json
client= OpenAI()


def get_stock_price(symbol):
    """Get the current stock price for given symbol"""
    stock_info = {
        "symbol": symbol,
        "price": "1000"
    }
    return json.dumps(stock_info)


tools = [
    {
        "type":"function",
        "name": "get_stock_price",
        "description": "Get the current stock price for the given symbol",
        "parameters": {
            "type": "object",
            "properties": {  
                "symbol": {
                    "type": "string",
                    "description": "The symbol, e.g. AAPL",
                }
            },
            "required": ["symbol"],
        },
    }
]


response= client.responses.create(
    model="gpt-5.4-mini",
    input="Get stock symbol and price for Apple Inc",
    tools=tools
)

tool_call= response.output[0]
args = json.loads(tool_call.arguments)
result=get_stock_price(args["symbol"])
print(result)
```
