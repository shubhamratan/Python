```python
string1=str(input("Enter any String : "))
count=0
length=len(string1)
string2=string1.lower()
for i in range(0,length):
    if string2[i]=='a' or string2[i]=='e' or string2[i]=='i' or string2[i]=='o' or string2[i]=='u':
        count+=1

print(f"The total number of vowels in the string {string1} is :", count)
```

```python
string3=str(input("Enter any string : "))
string4=string3.lower()
reversed_string=string4[::-1]
if string4==reversed_string:
    print(f"The string {string3} is a pallindrome")
else:
    print(f"The string {string3} is not a pallindrome")
```
