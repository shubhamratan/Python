```python
class Person():
    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    def __str__(self):
        return (f"{self.name}, {self.age} years old")
    
    def __repr__(self):
        return (f"Person name={self.name}, age={self.age}")

person=Person("Shubham",29)
print(person)
print(repr(person))
```

```python
class Person():
    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    def __str__(self):
        return (f"{self.name}, {self.age} years old")
    
    def __repr__(self):
        return (f"Person name={self.name}, age={self.age}")
    
    def __len__(self):
        return len(self.name)
```

```python
person=Person("Shubham",29)
print(person)
print(repr(person))
print(len(person))
```
