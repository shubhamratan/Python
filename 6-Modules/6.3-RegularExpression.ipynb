```python
import re

sentence ="The quick brown fox jumps over the lazy dog."

# Search for a pattern

match=re.search("jumps", sentence)
if match:
    print("Match Found")
    print("Start Index:", match.start())
    print("End Index:", match.end())

```

```python
# Find all the occurences of a pattern

matches= re.findall("the",sentence, re.IGNORECASE)
print("Matches:", matches)
```

```python
# Replace all the occurences of a pattern

new_sentence= re.sub("fox","cat",sentence)
print("New Sentence: ", new_sentence)
```

```python
# Compile a regex for efficiency (if used multiple times)

pattern=re.compile(r"\b\w+\b") # Matches the whole word
words=pattern.findall(sentence)
print("Words: ",words)
```
