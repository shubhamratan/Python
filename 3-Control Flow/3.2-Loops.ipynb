```python
## While Loop

count=0

while count<5:
    print(count)
    count=count+1
    
```

```python
count=0

while count%2==0:
    print(count)
    count=count+1
```

```python
## Loop Control Statement

## Break - The break statement exits the loop prematurly

for i in range(10):
    if i==5:
        break
    if i!=5:
        print(i)
```

```python
## Continue - The continue statement skips the current iteration and continues with the next.

for i in range(10):
   if i%2==0:
     continue
   print(i)
```

```python
## Pass - The pass statement is a null operation and it does nothing.

for i in range(5):
    if i==3:
         pass
    print(i)
```

```python
## Nested Loop - a loop inside a loop

for i in range(5):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
## Examples - Calculate the sume of first n natural numbers using a while and for loop

## While Loop
 
n=int(input("Enter any natural number"))
sum=0
count=1

while count<=n:
    sum=sum+count
    count=count+1

print("The of first",n,"natural numbers is:",sum)
```

```python
## For Loop

n=int(input("Enter any natural number"))
sum=0
count=1

for i in range(n):
    sum=sum+count
    count=count+1

print("The of first",n,"natural numbers is:",sum)
```
