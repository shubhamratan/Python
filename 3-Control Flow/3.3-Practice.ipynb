```python
user_age=29
guest_age=int(input("Enter your age:"))
if user_age==guest_age:
    print(f"We both are {user_age} years old")
elif user_age>guest_age:
    age_diff=user_age-guest_age
    if age_diff<2:
        print(f"I am only a year older than you")
    else:
        print(f"I am {age_diff} years older than you")
else:
    age_diff=guest_age-user_age
    if age_diff<2:
        print(f"You are only a year older than me")
    else:
        print(f"You are {age_diff} years older than me")
    
```
