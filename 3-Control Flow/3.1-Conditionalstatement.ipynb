```python
## If Statement
age=30

if age>=18:
    print("You are an adult")
```

```python
## Else Statment
age=16

if age>=18:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")
```

```python
## Elif Statment
age=16

if age<13:
    print("You are a child")
elif age<18:
    print("You are a teenager")
else:
    print("You are an adult")
```

```python
## Match Case Statement

x=int(input("Enter a number between 1 to 5: "))

match x:
    case 1:
        print("You have won a camera")
    case 2:
        print("You have won a car")
    case 3:
        print("You have won a bike")
    case _:
        print("Better luck next time")
```

```python
## Nested Conditional Statement

num=int(input("Enter any number"))

if num>0:
    print("Number is postive integer")
    if num%2==0:
        print("Number is even")
    else:
        print("Number is odd")

else:
    print("Number is zero or negative integer")
```

```python
## Simple Calculator Program

num1 = float(input("Enter 1st number:"))
num2 = float(input("Enter 2nd number:"))

operation = input("Enter operation (+, -, *, /): ")

if operation == '+':
    result = num1 + num2
elif operation == '-':
    result = num1 - num2
elif operation == '*':
    result = num1 * num2
elif operation == '/':
    if num2 != 0:
      result = num1 / num2
    else:
      result = "Error ! Division by zero."
else:
    result ="Invalid Operation" 

print(result)
```
