```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<p>Hello, World!</p>"

app.run(debug=False, use_reloader=False)
```

```python
from flask import Flask, render_template

app1 = Flask(__name__)

@app1.route("/")
def hello_world():
    return render_template("index.html")

app1.run(debug=True, use_reloader=False)
```
