```python
from flask import Flask, render_template, request 

app = Flask(__name__)

@app.route("/feedback", methods=["POST","GET"])

def feedback():
    if request.method == "POST":
        name=request.form.get("username")
        message=request.form.get("message")

        return render_template("thankyou.html", user=name, message=message)
    
    return render_template("feedback.html")

app.run(debug=True, use_reloader=False)
```

```python
from flask import Flask, render_template, request, url_for, flash, redirect

app = Flask(__name__)
app.secret_key="my-secret-key"

@app.route("/", methods=["GET", "POST"])

def feedbackform():
    if request.method == "POST":
        name=request.form.get("name")
        if not name:
            flash("Name cannot be empty")
            return redirect(url_for("feedbackform"))
        flash(f"Thanks {name}, your form was submitted")
        return render_template("thankyoupage.html", user=name)
    
    return render_template("feedbackform.html")

@app.route("/thankyoupage")
def thankyou():
    return render_template("thankyoupage.html")

app.run(debug=True, use_reloader=False)
```
