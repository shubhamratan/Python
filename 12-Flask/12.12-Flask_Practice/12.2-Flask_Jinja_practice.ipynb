```python
from flask import Flask, render_template

app = Flask(__name__)

@app.route("/")

def student_profile():
    return render_template(
        "profile.html", 
        name="Shubham",
        is_topper=True,
        subjects=["Maths","Science","History"]
        )

app.run(debug=True, use_reloader=False)
```
