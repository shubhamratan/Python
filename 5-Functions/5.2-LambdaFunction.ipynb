```python
#Syntax

lambda arguments: expression
```

```python
def addition(a,b):
    return a+b
```

```python
addition(2,3)
```

```python
addition = lambda a,b : a+b
type(addition)
```

```python
addition(2,3)
```

```python
def even(num):
    if num%2==0:
        return True

even(24)
```

```python
even1 = lambda a:a%2==0
print(even1(16))
```

```python
def addition1(x,y,z):
    return(x+y+z)

addition1(12,13,14)
```

```python
addition2 = lambda a,b,c: a+b+c
print(addition2(12,13,14))
```
