```python
def factorial(n):
    if n==0:
        return 1
    return n*factorial(n-1)

print(factorial(5))
```

```python
def finonacci(n):
    if n==0 or n==1:
        return n
    else:
        return (finonacci(n-2)+finonacci(n-1))
    

print(finonacci(7))
```

```python
def sum_of_digits(n):
    if n==0:
        return 0
    return n%10 + sum_of_digits(n//10)

print(sum_of_digits(67679))
```

```python
def multiply(a,b):
    '''
    This function multiplies two number
    
    Parameters:
      a: a is the first number
      b: b is the second number
    
    Return:
      The product of a and b
    '''
    return a*b

print(multiply(5,6))
```

```python
def safe_divide(a,b):
    if b==0:
        return f"Cannot divide by {b}"
    else:
        return a/b
    
print(safe_divide(30,10))
print(safe_divide(3,0))
```

```python
def sum(*args):
    total=0
    for item in args:
        total+= item
    return(total)


print(sum(342,2,7,9,10))
```
