```python
def square(a):
    return a**2

print(square(4))
```

```python
numbers=[1,2,3,4,5,6,7,8]
list(map(square,numbers))
```

```python
## Lambda Function with Map
numbers=[1,2,3,4,5,6,7,8]

list(map(lambda x:x**2,numbers))
```

```python
## Map Multiple Iterable

num1=[1,2,3]
num2=[4,5,6]
added_list=list(map(lambda a,b: a+b,num1,num2))
print(added_list)
```

```python
# Use map to convert strings to integers

str_numbers=['1','2','3','4','5','6']
int_numbers=list(map(int,str_numbers))
print(int_numbers)
```

```python
words=['apple','banana','cherry']
upper_word=list(map(str.upper,words))
print(upper_word)
```

```python
 # Map Function to a dictionary

def get_name(person):
    return person['name']

people=[
    {'name':'Shubham', 'age': 29},
    {'name':'Raunak', 'age': 32},
    {'name':'Ashank','age':20}
]

list(map(get_name,people))
```
