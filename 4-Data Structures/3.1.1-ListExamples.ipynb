Example 1. Manage a to do list

 . Create a to do list to keep track of tasks


```python
to_do_list=["Buy Groceries", "Clean the house","Pay Bills",]

## Adding to task

to_do_list.append("Morning Walk")
to_do_list.append("Schedule a meeting")

## Removing a completed task

to_do_list.remove("Clean the house")

## Checking if the task in the list

if "Pay Bills" in to_do_list:
    print("Don't forget to pay the utilities bill")

print("To do list remaining")
for task in to_do_list:
    print(f"-{task}")
```

Organizing Studnets Grades

Create a list to store and calculate average grades for students

```python
## Organizing Students Grades

grades=[80,90,78,89,65]
grades.append(81)

## Calculating the average grade

average_grade = sum(grades)/len(grades)
print(f"Average Grades: {average_grade:.2f}")

## Finding the highest and lowest grade

highest_grade = max(grades)
lowest_grade = min(grades)

print(f"Highest Grade: {highest_grade}")
print(f"Lowest Grade: {lowest_grade}")
```

```python
# Managing an inventory

inventory =["Pen","Pencil","Paper","Diary","Sharpner"]

## Adding a new item
inventory.append("Eraser")

## Removing an item
inventory.remove("Pen")

## Checking if an item is in stock
item= "Paper"

if item in inventory:
    print(f"{item} is in stock")
else:
    print(f"{item} is out of stock")

## Printing the inventory list

print("Inventory list:")
for item in inventory:
    print(f"- {item}")
```
