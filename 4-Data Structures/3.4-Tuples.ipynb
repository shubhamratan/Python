```python
## Packing & Unpacking
## Packing
packed_tuple=1,"hello",3.14
print(packed_tuple)
```

```python
## Unpacking

a,b,c=packed_tuple
print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuple
## Nested List

lst=[[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,'c']]
lst[0][0:4:2]
```

```python
lst=[[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,'c')]
lst[2][0:3]
```

```python
nested_tuple=((1,2,3),('a','b','c'),("Hello","world","I"))

## Access the elements inside the tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## Iterating over a nested tuple

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
    
```
