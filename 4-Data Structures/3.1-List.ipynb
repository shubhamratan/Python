## List Comprehension

Basic Syntax                 [Expression for item in iterable]

with condtional logic        [Expression for item in iterable if condition]

Nested List Comprehension    [Expression for item1 in iterable1 if item2 in iterable 2]

```python
### Basic List Comprehension

square =[num**2 for num in range(10)]
print(square)
```

```python
## List Comprehension with Condition

lst=[]
for i in range(10):
    if i%2==0:
        lst.append(i)

print(lst)
```

```python
[i for i in range(10) if i%2==0]
```

```python
## Nested List Comprehension

lst1=[1,2,3,4]
lst2=['a','b','c','d']

pair=[[i,j] for i in lst1 for j in lst2]

print(pair)
```

```python
## List Comprehensions with function call

words = ["Hello", "World","python","List","Comprehension"]
lenghts =[len(word) for word in words]
print(lenghts)
```
