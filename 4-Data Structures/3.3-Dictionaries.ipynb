```python
empty_dict=dict()
empty_dict
```

```python
student={"name":"Shubham","age":30,"Grade":'A'}
print(student)
print(type(student))
```

```python
## Single key is always used

student={"name":"Shubham","age":30,"name":24}
print(student)
```

```python

student={"name":"Shubham","age":30,"grade":'A'}
print(student)
```

```python
## Accessing Dictionaries Elements

print(student['grade'])
print(student['name'])

## Accessing Dictionaries Elements using get() method

print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not Available"))
```

```python
## Modyfying Dictionary Elements
## Dictionary are mutable, so you can add, update or delete elements
print(student)
```

```python
student["age"]=33  ## update the value for a key
print(student)
student["address"]="India"  ## added a new key and value 
print(student)
```

```python
del student['grade'] ## Delete the key value pair
print(student)
```

```python
## Common Dictionaries Methods in Pythons

keys=student.keys()  ## get all the keys
print(keys)
values=student.values() ## get all the values
print(values)

items=student.items()  ## get all the key value pairs
print(items)
```

```python
## Shallow Copy

student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="Shubham2"
print(student)
print(student_copy)
```

```python
student_copy1=student.copy()
print(student_copy1)
print(student)
```

```python
student["name"]="Shubham3"
print(student_copy1)
print(student)
```

```python
## Iterating Over Dictionaries
## You can use loops to iterate over dictionaries, keys, value, or items

## Iterating over keys

for keys in student.keys():
    print(keys)
```

```python
## Iterating over values

for values in student.values():
    print(values)
```

```python
## Iterating over key value pair

for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested Dictionaries - A dictionary inside a dictionary

students={
    "student1":{"name":"Shubham","age":30},
    "student2":{"name":"Happy","age":28}
}
print(students)
```

```python
## Accessing Nested Dictionaries Elements

print(students['student1']['name'])
print(students['student2']['age'])
```

```python
students.items()
```

```python
## Iterating over nested Dictionaries

for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
    
```

```python
## Dictionary Comprehension

squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Conditional Dictionary Comprehension
even_squares={x:x**2 for x in range(10) if x%2==0}
print(even_squares)
```

```python
## Practical Examples

## Use a dictionary to count the frequency of elements in the  list

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}
for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## Merge 2 dictionaries in one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dic={**dict1,**dict2}
print(merged_dic)
```
