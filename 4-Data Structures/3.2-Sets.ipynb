```python
## Create a set

my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set()  ## Directly using the set class
my_set=set([1,2,3,4,5,6])
print(type(my_empty_set))
print(my_set)
```

```python
my_set=set([1,2,3,4,5,6,2,3,6])
print(my_empty_set)
```

```python
## Basic Set operations
## Adding and Removing Elemnents

## Adding an element
my_set.add(7)
print(my_set)

## Even if you add an element twice
my_set.add(7)
print(my_set)
```

```python
## Removing an element 

my_set.remove(3)
print(my_set)
```

```python
## Removing an element which is not present in set
my_set.remove(10)
print(my_set)
```

```python
my_set.discard(10)  ## This will not throw an error
print(my_set)
```

```python
## pop method - It follows FIFO 

removed_element = my_set.pop() 
print(removed_element)
print(my_set)
```

```python
## Clear All members of the set

my_set.clear()
print(my_set)
```

```python
## Set Membership Test

my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## Mathematical Operation

set1={1,2,3,4,5,6}
set2={4,5,6,7,8}

## Union
union_set=set1.union(set2)
print(union_set)

## Intersection

intersection_set=set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8}

## Difference

print(set1.difference(set2))
```

```python
set1={1,2,3,4,5,6,7,8}
set2={5,6,7,8}

set1.difference_update(set2)

print(set1)
```

```python
## Symmetric Difference

set1.symmetric_difference(set2) 

```

```python
## Set Methods

set1={1,2,3,4,5}
set2={3,4,5}

## is subset

print(set1.issubset(set2))

print(set1.issuperset(set2))
```

```python
lst=[1,2,3,4,5,6,7,4,6,9]

set(lst)
```

```python
## Counting Unique words in text

text="Hi my name is Shubham Ratan"
words=text.split()

## Convert list of words to set to get unique words

unique_words=set(words)
print(unique_words)
print(len(unique_words))
```
