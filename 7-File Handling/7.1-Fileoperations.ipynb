```python
# Write to a file called John Doe.txt
# It should contain data about John Doe

f = open("John Doe.txt","w")

string='''
John Doe is a nice guy. He lives in New York City and he works with python
His favourite package is Pandas.
'''

f.write(string)

f.close()
```

```python
# Reading a Whole File using "with"
# No need to write f.close() as file is closed automatically by using with Statement

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
# Reading a file line by line

with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) ## strip removes the newline character
```

```python
# Writing a file(Overwriting)

with open('example.txt','w') as file:
    file.write('Hello World!\n')
    file.write('This is a new line.')
    
```

```python
# Writing a file without Overwriting


with open('example.txt','a') as file:
    file.write('\nAppend Operation\n')
    file.write('This is a new line.')
    
```

```python
# Writing a list of lines in file
lines=['First Line \n','Second Line \n','Third Line \n']
with open('example.txt','a') as file:
    file.writelines(lines)
    
```

```python
# Binary Files

# Writing to a binary file

data=b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
# Reading a binary file

with open('example.bin','rb') as file:
    content=file.read()
    print(content)
```

```python
# Read the content from a source text file and write it to a destination text file

with open('example.txt','r') as source_file:
    content=source_file.read()

with open('destination.txt','w') as destination_file:
    destination_file.write(content)
```

```python
# Read a text file and count the number of line, word and characters.

def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines=file.readlines()
        line_count=len(lines)
        word_count=sum(len(line.split())for line in lines)
        char_count=sum(len(line)for line in lines)
    return line_count,word_count,char_count

file_path='example.txt'
lines,words,characters= count_text_file(file_path)
print(f'Lines: {lines} , Words: {words}, Characters: {characters}' )    
```

```python
# Writing and then reading of file

with open('example.txt','w+') as file:
    file.write("Hello World\n")
    file.write("This is a new line \n")

    # Move the file cursor to the beginning
    file.seek(0)
    
#Read the content of the file

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```
