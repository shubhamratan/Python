```python
# Series 
import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series :\n",series)
print(type(series))
```

```python
#Create a series from Dictionaries

data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
print(type(series_dict))
```

```python
#Giving Different index based on our requirement
data = [10,20,30]
print(pd.Series(data))

index=['a','b','c']
pd.Series(data,index=index)
```

```python
#DataFrame
# Creating a DataFrame from a dictionaries of list

data={
    'Name':['Shubham','Raunak','Rachit'],
    'Age':[30,35,20],
    'City':['Bangalore','Patna','Delhi']
}

df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
#
import numpy as np

np.array(df)
```
