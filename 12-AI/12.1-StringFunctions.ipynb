```python
import re

def to_slug(text):
    # Convert to lowercase
    text = text.lower()
    
    # Replace anything that's not a letter or number with a space
    text = re.sub(r'[^a-z0-9]+', ' ', text)
    
    # Strip leading/trailing spaces
    text = text.strip()
    
    # Replace spaces with dashes
    slug = text.replace(' ', '-')
    
    return slug

print(to_slug(" hey you are good "))
print(to_slug("Hello@World!! Python#Rocks"))
print(to_slug("  Python   is   awesome  "))
print(to_slug("Clean-Code!!! Is#Important"))
print(to_slug("123 Go Fast!!!"))
print(to_slug("Make $$$ Money Quickly"))
print(to_slug("  MixedCASE String With Spaces "))
```
